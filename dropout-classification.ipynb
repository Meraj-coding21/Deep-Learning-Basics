# *Training and visualising with different dropout_rates*
ANN is prone to overfitting. If there is no dropout (switching off some random nodes) then there is a chance of overfitting. if we increase the dropout rate more than 0.5 there is a chance of underfitting. so we need to find a particular balance of dropout rates

```python
dropout_rates = [0, 0.2, 0.3, 0.5]
plt.figure(figsize=(15,20))

for i, dropout_rate in enumerate(dropout_rates):

    print(f"Training for dropout_rate: {dropout_rate}")
 
    model = Sequential()
    model.add(Dense(128,input_dim=X.shape[1],activation='relu'))

    if(dropout_rate > 0):
        model.add(Dropout(dropout_rate))

    model.add(Dense(128,activation='relu'))

    if(dropout_rate > 0):
        model.add(Dropout(dropout_rate))

    model.add(Dense(1, activation='sigmoid'))

    adam = Adam(learning_rate=0.01)

    model.compile(loss='binary_crossentropy',optimizer=adam, metrics=['accuracy'])

    history = model.fit(X,y, epochs = 500, validation_split=0.2, verbose=False)


    #decision regions
    plt.subplot(len(dropout_rates),2,2*i+1)
    plot_decision_regions(X,y,clf=model,legend=2)
    plt.xlim(-0.7,0.7)
    plt.ylim(-0.8,0.8)
    plt.title(f'Decision Region (Dropuout Rate: {dropout_rate})')

    #loss curve
    plt.subplot(len(dropout_rates),2,2*i+2)
    plt.plot(history.history['loss'], label = 'Training Loss')
    plt.plot(history.history['val_loss'], label = 'Validation Loss')
    plt.title(f'Loss (Dropout Rate : {dropout_rate})')
    plt.legend()


plt.tight_layout()
plt.show()
```
