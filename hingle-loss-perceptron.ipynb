```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
(X,y) = make_classification ( n_samples = 100, n_features = 2, n_classes = 2, random_state = 41, 
                              n_clusters_per_class = 1, class_sep = 15, n_informative = 1, n_redundant = 0,
                              hypercube = False)
```

```python
plt.figure ( figsize = (10, 6))
plt.scatter ((X[:,0]),(X[:,1]), c = y, cmap ="winter", s = 100)
```

```python
def perceptron(X,y):
    epoch = 1000
    w1=w2=b=1
    learning_rate = 0.1
         
    for j in range(epoch):
        for i in range(len(X)):
            z = w1*X[i][0] + w2*X[i][1] + b
            if (y[i]*z < 0):
                w1 = w1 + learning_rate * y[i]*X[i][0]
                w2 = w2 + learning_rate * y[i]*X[i][1]
                b = b + learning_rate * y[i]
    return w1,w2,b
```

```python
w1,w2,b = perceptron(X, y)
```

```python
print(w1)
print(w2)
print(b)
```

```python
m = -(w1/w2)
c = -(b/w2)

print (m,c)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
plt.figure ( figsize = (10, 6))
plt.plot(x_input, y_input, color = 'red', linewidth = 3)
plt.scatter ((X[:,0]),(X[:,1]), c = y, cmap ="winter", s = 100)
plt.ylim(-4,4)
```
